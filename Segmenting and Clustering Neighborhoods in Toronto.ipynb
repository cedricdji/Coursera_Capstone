<b>2 - Scraping the List of postal codes</b>

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
```

```python
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

Data preprocessing and cleaning

```python
# Get names of indexes for which column Borough has value "Not assigned"
indexNames = df[ df['Borough'] =='Not assigned'].index

# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)

df.head(10)
```

<strong>If a cell has a borough but a Not assigned  neighborhood, then the neighborhood will be the same as the borough.</strong>

```python
df.loc[df['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = df['Borough']
df.head(10)
```

<strong>combined into one row with the neighborhoods separated with a comma</strong>

```python
Rows_same = df.groupby(['Postal Code','Borough'], sort=False).agg( ', '.join)
df_new=Rows_same.reset_index()
df_new.head(15)
```
